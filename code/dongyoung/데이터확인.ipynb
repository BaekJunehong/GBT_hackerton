```python
from collections import Counter

def get_most_common_keywords(df, category):
  
  # Extract rows based on category
  df_category = df[df['분류'].str.contains(category)]
  
  # Split keywords and count occurrences
  keywords = df_category['키워드'].str.split(',').explode()
  most_common_keywords = Counter(keywords).most_common()
  
  return most_common_keywords

# Example usage
most_common_keywords = get_most_common_keywords(df, '스포츠')
most_common_keywords
```

```python
# 각 인덱스마다 2번 이상 반복되는 키워드를 추출하는 함수

def get_repeated_keywords(df, row_index):
    # Get the keywords from the specified row
    first_row_keywords = df.iloc[row_index]['키워드']
    
    # Split the keywords and count occurrences
    keywords_list = first_row_keywords.split(',')
    keyword_counts = Counter(keywords_list)
    
    # Remove keywords that are not repeated
    filtered_keywords = [keyword for keyword in keywords_list if keyword_counts[keyword] > 1]
    
    # Remove duplicate keywords
    unique_filtered_keywords = list(dict.fromkeys(filtered_keywords))
    
    category = df.iloc[row_index]['분류']
    
    return category, unique_filtered_keywords

# 사용 예시
repeated_keywords = get_repeated_keywords(df, 3001)  # 행 인덱스 입력
print(repeated_keywords)
```

```python
# 각 인덱스마다 키워드 중 1번만 반복되는 키워드를 추출하는 함수

def get_one_keyword(df, row_index):
    # Get the keywords from the specified row
    first_row_keywords = df.iloc[row_index]['키워드']
    
    # Split the keywords and count occurrences
    keywords_list = first_row_keywords.split(',')
    keyword_counts = Counter(keywords_list)
    
    # Remove keywords that are not repeated
    filtered_keywords = [keyword for keyword in keywords_list if keyword_counts[keyword] == 1]
    
    # Remove duplicate keywords
    unique_filtered_keywords = list(dict.fromkeys(filtered_keywords))
    
    category = df.iloc[row_index]['분류']
    
    return category, unique_filtered_keywords

# 사용 예시
repeated_keywords = get_one_keyword(df, 3001)  # 행 인덱스 입력
print(repeated_keywords)
```

```python
def extract_top_keywords_by_group(df, group_column, keyword_column, top_n=5):
    # 그룹화
    grouped = df.groupby(group_column)
    
    top_keywords = {}
    
    for group_name, group_df in grouped:
        # 모든 키워드 결합
        all_keywords = group_df[keyword_column].dropna().str.cat(sep=',')
        
        # 키워드 분리 및 빈도 계산
        keywords_list = all_keywords.split(',')
        keyword_counts = Counter(keywords_list)
        
        # 가장 많이 등장하는 키워드 추출
        if keyword_counts:
            top_keywords[group_name] = [keyword for keyword, count in keyword_counts.most_common(top_n)]
        
        # 각 그룹의 키워드를 줄바꿈으로 구분하여 출력
        print(f"Group: {group_name}, Keywords: {top_keywords[group_name]}\n")
    
    return top_keywords

# 사용 예시
top_keywords = extract_top_keywords_by_group(df, '분류', '키워드', top_n=5)
print(top_keywords)
```

'용인시', '용인' 키워드가 모든 분류에서 높은 빈도로 사용되는 것을 발견 -> 분류에 중요하지 않은 키워드라 판단하여 제거하면 좋을 듯

```python
from collections import Counter

def extract_single_character_keywords(df, keyword_column):
    """
    데이터프레임의 각 행에서 한 글자인 키워드를 추출합니다.
    """
    single_char_keywords = []
    
    for index, row in df.iterrows():
        if pd.notna(row[keyword_column]):
            keywords_list = row[keyword_column].split(',')
            single_char_keywords.extend([keyword for keyword in keywords_list if len(keyword) == 1])
    
    return single_char_keywords

def count_single_character_keywords(df, keyword_column):
    """
    데이터프레임의 각 행에서 한 글자인 키워드를 추출하고, 빈도를 계산하여 딕셔너리 형태로 반환합니다.
    """
    single_char_keywords = extract_single_character_keywords(df, keyword_column)
    keyword_counts = Counter(single_char_keywords)
    return dict(keyword_counts)

single_char_keyword_counts = count_single_character_keywords(df, '키워드')
print(single_char_keyword_counts)
```

한글자인 단어는 전부 drop처리

```python
# 한국어 불용어 리스트
stopwords = ['초', '중', '고', '농', '민', 
             '폭', '군', '시', '게', '등', 
             '남', '동',  '여', '콩', '재', 
             '가', '인', '존', '면', '관', 
             '분', '통', '착', '탈', '입', '소', '신']

def remove_stopwords(keywords_list, stopwords):
    """
    주어진 키워드 리스트에서 불용어를 제거합니다.
    """
    return [keyword for keyword in keywords_list if keyword not in stopwords]

def process_row_keywords(row_keywords, stopwords):
    """
    주어진 행의 키워드 문자열을 분리하고 불용어를 제거합니다.
    """
    keywords_list = row_keywords.split(',')
    return remove_stopwords(keywords_list, stopwords)

def remove_stopwords_from_dataframe(df, keyword_column):
    """
    데이터프레임의 각 행마다 키워드 컬럼에서 불용어를 제거합니다.
    """
    for index, row in df.iterrows():
        if pd.notna(row[keyword_column]):
            processed_keywords = process_row_keywords(row[keyword_column], stopwords)
            df.at[index, keyword_column] = ','.join(processed_keywords)
    return df
```

```python
df = remove_stopwords_from_dataframe(df, '키워드')
print(df)
```

숫자-한국어, 영어-한국어를 혼용한 단어 확인 및 제거

```python
import re
```

```python
text = "여기 A씨, 11명과 hello-하이, 123abc-안녕하세요가 혼합된 예시입니다."

# 혼합된 단어 추출을 위한 정규표현식
mixed_words = re.findall(r'\b[A-Za-z]+[가-힣]+\b|\b[가-힣]+[A-Za-z]+\b|\b[0-9]+[가-힣]+\b|\b[가-힣]+[0-9]+\b', text)

print(mixed_words)
```

```python
def find_mixed_language_keywords(df, keyword_column):
    """
    데이터프레임의 각 행에서 숫자, 한글, 영어가 혼합된 단어를 추출합니다.
    """
    pattern = re.compile(r'\b[A-Za-z]+[가-힣]+\b|\b[가-힣]+[A-Za-z]+\b|\b[0-9]+[가-힣]+\b|\b[가-힣]+[0-9]+\b')
    
    mixed_words = []
    
    for index, row in df.iterrows():
        if pd.notna(row[keyword_column]):
            keywords_list = row[keyword_column].split(',')
            mixed_words.extend([keyword for keyword in keywords_list if pattern.search(keyword)])
    
    # 빈도 계산
    keyword_counts = Counter(mixed_words)
    return dict(keyword_counts)

# 혼합된 단어와 그 빈도를 딕셔너리 형태로 반환
mixed_language_keyword_counts = find_mixed_language_keywords(df, '키워드')

# 딕셔너리를 데이터프레임으로 변환
mixed_language_keyword_df = pd.DataFrame(list(mixed_language_keyword_counts.items()), columns=['단어', '빈도'])

# 테이블 형태로 출력
mixed_language_keyword_df
```
