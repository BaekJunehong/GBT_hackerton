```python
remove_keywords = list(set(remove_keywords))
def remove_keywords_from_keywords_column(df, remove_keywords):
    # 키워드 열에서 각 키워드를 삭제하는 함수
    def filter_keywords(keywords):
        if pd.isna(keywords):  # 키워드가 없으면 처리하지 않음
            return keywords
        keywords_list = keywords.split(',')  # 쉼표로 구분된 키워드 리스트
        filtered_keywords = [kw for kw in keywords_list if kw not in remove_keywords]  # 삭제할 키워드를 제거
        return ','.join(filtered_keywords)  # 다시 쉼표로 구분된 문자열로 변환

    df['키워드'] = df['키워드'].apply(filter_keywords)  # 각 행의 키워드 열에 적용
    return df


train_df = remove_keywords_from_keywords_column(train_df, remove_keywords)
test_df = remove_keywords_from_keywords_column(test_df, remove_keywords)
```
